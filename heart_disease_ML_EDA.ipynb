```python
# distribution of labels in our data
sns.countplot(x=df['HeartDisease'])
plt.show()
```

```python
# separate numerical from categorical features
continuous_var = df.select_dtypes('float64').columns
print('continuous_var:', continuous_var)
categorical_var = df.select_dtypes('object').columns
print('categorical_var:', categorical_var)
```

```python
# multiple feature plots
i=1
plt.figure(figsize = (20,16))
for col in continuous_var:
  plt.subplot(2,2,i)
  sns.kdeplot(x=col,hue='HeartDisease',data=df)
  plt.xlabel(col)
  plt.ylabel("Count")
  i+=1
```

```python
# continuous feature box plots
i=1
plt.figure(figsize = (20,16))
for col in continuous_var:
  plt.subplot(2,2,i)
  sns.boxplot(y=col,x='HeartDisease',data=df)
  plt.xlabel(col)
  plt.ylabel("Count")
  i+=1
```

```python
sns.boxplot(y='PhysicalHealth', data=df)
```

```python
# compare the categorical variables
for col in categorical_var[1:]:
  g = sns.catplot(x=col, col='HeartDisease', kind='count', data=df, sharey=False)
  g.set_xticklabels(rotation=60)
```
